**DNA transcription**

Write a Python code that simulates the process of transcription from DNA to RNA, but this time, you must use dictionaries instead of using other methods

**Example**:


CCGTAGCGTAGC ---> GGCAUCGCAUCG

```python
seq = 'GCATCTAGTCGTCGATCTCGTAGTCTATCGATCTAGCCTGACT'
```

```python
dna_to_rna = {'A':'U','C':'G','G':'C','T':'A'}
```

```python
rna_seq = ''
for nucleotide in seq:
  rna_seq = rna_seq + dna_to_rna[nucleotide]
```

```python
print(f'The RNA sequence is {rna_seq}')
```
