#Biomolecule classifier

Write a code that takes a molecule as input and classify it as DNA or RNA by using conditional statements


For example:



**GCGUAGCUACGUCUAG** ----> This molecule is RNA

**CGTAGCGATGATCGAC** ----> This molecule is DNA


Input

```python
molecule = str(input('Type the molecule: ')).upper()
```

```python
molecule
```

Conditional

```python
if 'T' in molecule:
  print('This is a DNA molecule')
elif 'U' in molecule:
  print('This is RNA molecule')
else:
  print('Unknown type of molecule')
```
