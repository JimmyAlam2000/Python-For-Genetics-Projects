
**Extracting the AT/CG content**


Write a function that extracts the proportion of A-T or C-G in a DNA sequence. The function must have two parameters:

1- Sequence, i.e. the sequence to be analyzed

2- The base (A,T,C,G) of the base-pair to be analyzed (A-T, or C-G)

```python
def basePairProportion(dna,base):
  baseProportion = dna.count(base)
  basePairing = {'A':'T','T':'A','C':'G','G':'C'}
  secondBase = basePairing[base]
  secondBaseProportion = dna.count(secondBase)
  basePairAmount = baseProportion + secondBaseProportion
  percentage = (basePairAmount/len(dna))*100
  return percentage
```

```python
dna1 = 'CGATCGATCGATCTGCTGATTAGTGCGATCGTAGTCGTATCGTATGCTAGTGCTATCAGCTCGT'
dna2 = 'CGTACGACTACGGACTATCATCGACTACTGATCGTTGTGTGTGTCGTACGATAGCTAGACTATG'
```

```python
basePairProportion(dna1,'A')
```
