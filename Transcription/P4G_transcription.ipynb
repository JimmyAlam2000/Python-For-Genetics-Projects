**DNA transcription**

Write a Python code that simulates the process of transcription from DNA to RNA

**Example**:


CCGTAGCGTAGC ---> GGCAUCGCAUCG

Sequence

```python
dna = str(input('Type the DNA sequence: '))
```

Conditionals

```python
rna = ''

for nucleotide in dna:
  if nucleotide == 'T':
    rna += 'U'
  else:
    rna += nucleotide
```

```python
print(f'The transcribed rna sequence is {rna}')
```
