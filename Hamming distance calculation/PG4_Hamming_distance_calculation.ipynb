#**Hamming distance**

The hamming distance of two strings can be understood as the amount of characters that are different between these two strings.


Given two DNA sequences of the same length, write a code that calculates the hamming distance between the two strings and the similarity (%) between them.

```python
a = 'CGATGCTGCATGCTGACTGCTAGC'
b = 'CGCTGCTCCATTCTGACGGCTAGC'
```

```python
count = 0
difference = 0
while count < len(a):
  if a[count] != b[count]:
    difference = difference + 1
  count = count + 1

percentageOfDifference = (difference/len(a))*100
```

```python
difference
```

```python
print(f'The hamming distance between the two sequences is {difference} and the similarity is {100-percentageOfDifference} %')
```
