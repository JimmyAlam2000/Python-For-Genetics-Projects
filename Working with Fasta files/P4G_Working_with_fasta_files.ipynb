#Transcribing a DNA sequence from a fasta file

Go to some gene/genome repository and download some fasta file of some DNA sequence. Import the file to the COLAB environment and write a code that transcribes the sequence and generates a .txt file with the sequence transcribed

```python
with open('/content/sequence.fasta','r') as sarscov2:
  sequence = []
  count = 0
  for line in sarscov2:
    if count > 0:
      sequence.append(line)
    count += 1
```

```python
sequence2 = ''.join(sequence)
```

```python
sequence2
```

```python
dna_to_rna = {'A':'U','T':'A','C':'G','G':'C'}
rnaSeq = ''
for character in sequence2:
  if character in dna_to_rna.keys():
    rnaSeq += dna_to_rna[character]
  else:
    rnaSeq += character
```

```python
rnaSeq
```

```python
#Creating a txt file for the RNA sequence
with open('rnaSeq.txt','w') as rna:
  rna.write(rnaSeq)
```
