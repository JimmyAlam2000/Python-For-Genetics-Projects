**Generating a random DNA sequence**


Write a function that generates a random DNA sequence. This function must receive the sequence length as the only parameter

For example:

generateDNAseq(10) ---> AGTGCGGATC

generateDNAseq(13) ---> AGTCGAGGCTGAC

```python
import random
```

```python
def generateDNAseq(length):
  nucleotides = ('A','T','C','G')
  randomSeq = ''
  for nucleotide in range(length):
    randomSeq += random.choice(nucleotides)
  return randomSeq
```

```python
generateDNAseq(50)
```
