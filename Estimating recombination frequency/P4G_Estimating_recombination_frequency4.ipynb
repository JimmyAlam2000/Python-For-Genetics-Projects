Write a program that estimates gene distances based on the frequencies of the recombinant genotypes and the non-recombinant genotypes for examples:

AB: 132 (non-recombinant)

ab: 145 (non-recombinant)

aB: 11 (recombinant)

Ab: 13 (recombinant)

```python
rec1 = int(input('Number of recombinant genotypes 1:  '))
rec2 = int(input('Number of recombinant genotypes 2:  '))
nonrec1 = int(input('Number of non-recombinant genotypes 1:  '))
nonrec2 = int(input('Number of non-recombinant genotypes 2:  '))
```

```python
nonrec2
```

```python
RF = 100*(rec1+rec2)/(rec1+rec2+nonrec1+nonrec2)
```

```python
RF
```

```python
print('The recombination frequency is estimated to be',round(RF,3),'%')
```
