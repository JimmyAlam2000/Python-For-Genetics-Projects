# Calculating allele frequencies

Write a Python code that asks the user the number of individuals with the genotypes pp, pq and qq in the population and then returns the frequencies of the two alleles (p and q) in the population.


```python
pp = int(input('Number of pp genotypes: '))
qq = int(input('Number of qq genotypes: '))
pq = int(input('Number of pq genotypes: '))
```

```python
numberOfIndividuals = pp + qq + pq
```

```python
numberOfIndividuals
```

```python
#Frequencies of the alleles p and q
freqP = (2*pp + pq)/(numberOfIndividuals*2)
freqQ = (2*qq + pq)/(numberOfIndividuals*2)
```

```python
freqQ+freqQ
```

```python
print('The respective frequencies of all the alleles p and q are',round(freqP,2),'and',round(freqQ,2),'respectively!')
```
