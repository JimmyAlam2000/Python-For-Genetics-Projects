# Position of start and termination codons
Write a code that:

a) returns the position where the start and termination codons begin

b) extracts the ORF (open reading frame) of the RNA sequence

Tips: The start codon is AUG and the termination codon may be UGA, UAA and UAG

Example:

GGCCCGAUGGGAGGAGGGCCCUUGGGAUGAUUUGACCGA

start codon: position 6 termination codon: position 32

ORF: AUGGGAGGAGGGCCCUUGGGAUGAUUUGA

```python
rna = 'AAAAUGUUUCCCGCGGCGGCGUGAUUUAGCGAC'
```

```python
start = rna.find('AUG')
termination = rna.find('UGA')

print(f'The positions of the start and termination codons are {start} and {termination}, respectively')
```

```python
orf = rna[start:termination+3]
```

```python
print(f'ORF: {orf}')
```
